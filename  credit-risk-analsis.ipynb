# Steps involved in model building

1. Reqirment gathering or understand the problem
2. Data gathering( collecting the data)
3. Data cleaning (data pre processing) ---> handling null, replace, level incoder, rename,add,drop;
4. sampling ( divide your data into Train data and test data)
5. Build the model on the Train data
6. evaluate (judge the performance) the model on test data
7. run your model again if needed for any above step to improve the model

* Testing should not be done on the same data which we use to train data because getting it correct is high( high accuracy )
* Data is divided in two parts ( Train & Test) after then further dived into train_x & test_x dataframe

```python
cr = pd.read_csv(r"C:\Users\admin\Downloads\CreditRisk.csv")
cr
```

```python
cr.head(5)
```

```python
cr = cr.drop(['Loan_ID'] , axis =1)
```

```python
cr.Gender.fillna('Male', inplace = True)
cr.Married.fillna('Yes', inplace = True)
cr.Dependents.fillna(0, inplace = True)
cr.Self_Employed.fillna('No', inplace = True)
cr.LoanAmount.fillna(cr.LoanAmount.mean() ,inplace = True)
cr.Loan_Amount_Term.fillna(cr.Loan_Amount_Term.mean() ,inplace = True)
cr.Credit_History.fillna( 0 ,inplace = True)  
```

```python
from sklearn.preprocessing import LabelEncoder   # required import  and data should not contain nulls in data
le = LabelEncoder()
```

```python
cr.Gender = le.fit_transform(cr.Gender)
cr.Married = le.fit_transform(cr.Married)
cr.Education = le.fit_transform(cr.Education)
cr.Self_Employed = le.fit_transform(cr.Self_Employed)
cr.Property_Area = le.fit_transform(cr.Property_Area)
cr.Loan_Status = le.fit_transform(cr.Loan_Status)
```

```python
cr[cr.select_dtypes(include=['object']).columns] = cr[cr.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```
